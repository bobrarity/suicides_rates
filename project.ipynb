```python
uzb_df = df[df['country'] == 'Uzbekistan']
```

```python
plt.figure(figsize=(10,4),dpi=200)
sns.barplot(uzb_df, x='age',y='suicides/100k pop',hue='sex')
plt.title('Suicides in Uzbekistan by age from 1990 to 2014');
```

```python
uzb_df['year'].unique()
```

```python
plt.figure(figsize=(10,4),dpi=200)
sns.barplot(uzb_df, x='age',y='suicides/100k pop',hue='sex')
plt.title('Suicides in Uzbekistan by age from 1990 to 2014')
plt.xticks(rotation=45);
```

```python
plt.figure(figsize=(10,4),dpi=200)
sns.barplot(uzb_df, x='year',y='suicides/100k pop',hue='sex')
plt.title('Suicides in Uzbekistan by year from 1990 to 2014')
plt.xticks(rotation=45);
```

```python
df.groupby('country')['suicides/100k pop'].max().sort_values(ascending=False).head(10)
```

```python
top_5_countries = df.sort_values('suicides/100k pop', ascending=False).drop_duplicates('country').head(10)
```

```python
plt.figure(figsize=(10,4),dpi=200)
sns.barplot(top_5_countries, x='age',y='suicides/100k pop',hue='sex')
plt.title('The countries with most suicides per 100k by age from 1990 to 2014');
```

```python
plt.figure(figsize=(10,4),dpi=200)
sns.barplot(top_5_countries, x='country',y='suicides/100k pop',hue='sex')
plt.title('The countries with most suicides per 100k by year from 1990 to 2014')
plt.xticks(rotation=90);
```

```python
total_suicides = df.groupby('country')['suicides_no'].sum().reset_index().sort_values('suicides_no',ascending=False)
```

```python
total_suicides.head()
```

```python
plt.figure(dpi=200)
sns.barplot(total_suicides, x='country',y='suicides_no')
plt.title('Total number of suicides in countries')
plt.xlabel('Countries')
plt.ylabel('Total number of suicides')
plt.xticks(rotation=90);
```
